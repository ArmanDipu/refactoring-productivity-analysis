```python
import cufflinks as cf
import pandas as pd
cf.go_offline()
```

```python
commit_sha_column_name = "commit_sha"
```

```python
code_churn_dataset = pd.read_csv("../Dataset/codeChurnDataset.csv", engine="pyarrow")
# code_churn_dataset
```

```python
def extract_sha_in_brief_form(sha_value:str) -> str:
    return sha_value[:4] + "..." + sha_value[-4:]
```

```python
commit_shas = code_churn_dataset["url"].map(lambda x: extract_sha_in_brief_form(x.split("/")[-1]))
code_churn_dataset[commit_sha_column_name] = commit_shas
code_churn_dataset
```

```python
code_churn_dataset.iplot(
    kind = "scatter",
    x = "total refactoring types",
    y = "Total Code Churn",
    mode="markers",
)
```

```python
code_churn_dataset[[
    "total refactoring types",
    "Total Code Churn",
    "commit_sha"
]].head(10).iplot(
    kind = "bar",
    barmode = "group",
    categories="commit_sha",
        
)
```
